## Unknown Frequency

Since we are dealing, in this section, with an unknown frequency, it is easily assumed that the phase is also unknown. 

Furthermore, for the sake of simplicity, we will assume the following conditions about the unknown frequency random variable $ f_d $:

1. $ f_d $ is either present under *both* hypotheses.

    From this, we can infer that a single transmitter sends a signal corresponding to one of the two hypotheses. 
    
    $ f_d $ represents a Doppler shift and/or a frequency offset between the transmitter and receiver.  
  
2. or $ f_d $ is present under *one* hypothesis and the other hypothesis represents noise only,  
3. and $ f_d $ is constant for the entire measurement interval $ T $.


These assumptions can be relaxed, but only at the cost of a significant increase in complexity.

### Signal Modelling

Using these assumptions, if we have $ k $ samples over the duration $ T $, the $ j $-th sample under hypothesis $ H_i $ is

$$
y_j = \alpha a_i(t_j) e^{j 2\pi (f_i + f_d)t_j + \phi_i(t_j) + \beta_i} + z_j, \quad i = 0, 1, \quad j = 1, \dots, k
$$

where 
- $ \alpha $ is the attenuation, 
- $ f_i $ is the *known* frequency of transmission under hypothesis $ H_i $, 
- $ f_d $ is the *unknown* frequency shift, 
- $ \beta_i $ is the unknown received phase.

Note that $ \alpha $ is possibly unknown.

Using vector notation for the $ k $ samples, we can write

$$
\vec{y} = \alpha \vec{u}_i \vec{v} e^{j \beta_i} + \vec{z}
$$

where

$$
\vec{u}_i = \{ u_{i1}, u_{i2}, \dots, u_{ik} \}^T = \{ a_i(t_j) e^{j 2\pi f_i t_j + \phi_i(t_j)} \}^T
$$

represents the components corresponding to the known frequency $ f_i $ and

$$
\vec{v} = \{ v_1, v_2, \dots, v_k \}^T = \{ e^{j 2\pi f_d t_j} \}^T
$$

corresponds to the unknown frequency random variable $ f_d $. 

Both $ \vec{u}_i $ and $ \vec{v} $ are column vectors. 

### Conditional Density Functions

If the noise is zero mean and white with a correlation matrix of $ \sigma_z^2 I $, the conditional density functions are

$$
p_i(\vec{y}|\alpha, \beta_i, f_d) = \frac{1}{(2\pi \sigma_z^2)^k} \exp \left( -\frac{1}{2\sigma_z^2} \left[ \vec{y} - \alpha \vec{u}_i \vec{v} e^{j \beta_i} \right]^T \left[ \vec{y}^* - \alpha \vec{u}_i^* \vec{v}^* e^{-j \beta_i} \right] \right)
$$

which can be expanded to

$$
\begin{align*}
p_i(\vec{y}|\alpha, \beta_i, f_d) 
&= \frac{1}{(2\pi \sigma_z^2)^k} \exp \left( -\frac{1}{2\sigma_z^2} \sum_{j=1}^k |y_j|^2 \right) \\
&\times \exp \left( \frac{\alpha}{\sigma_z^2} \left| \sum_{j=1}^k y_j u_{ij} v_j^* \right| \cos(\beta_i - \xi_i) \right) \\
&\times \exp \left( -\frac{\alpha^2}{2\sigma_z^2} \sum_{j=1}^k |u_{ij} v_j|^2 \right), \quad i = 0, 1
\end{align*}
$$

Consider the last term in this equation. Since

$$
|u_{ij} v_j|^2 = u_{ij} u_{ij}^* v_j v_j^* = |u_{ij}|^2 |v_j|^2
$$

and

$$
|v_j|^2 = (e^{j 2\pi f_d t_j})(e^{-j 2\pi f_d t_j}) = 1
$$

it follows that the last term is

$$
\exp \left( -\frac{\alpha^2}{2\sigma_z^2} \sum_{j=1}^k |u_{ij}|^2 \right) = \exp \left( -\frac{\alpha^2 \mathcal{E}_i}{\sigma_z^2} \right)
$$

Just as in previous sections, if the phase $ \beta_i $ is assumed uniform in the interval $ (0, 2\pi) $. 

We can average over the phase to produce the conditional density function

$$
\boxed{
\begin{align*}
p_i(\vec{y}|\alpha, f_d) 
&= \frac{1}{(2\pi\sigma_z^2)^k} \exp \left( -\frac{1}{2\sigma_z^2} \sum_{j=1}^k |y_j|^2 \right) \\
&\times I_0 \left( \frac{\alpha}{\sigma_z^2} \left| \sum_{j=1}^k y_j u_{ij}^* v_j^* \right| \right) \exp \left( -\frac{\alpha^2 \mathcal{E}_i}{\sigma_z^2} \right)
\end{align*}
}
$$

### Likelihood Ratio

The first two terms have no effect on the decision; the conditional likelihood ratio is

$$
\boxed{
L(\vec{y}|\alpha, f_d) = \frac{I_0 \left( \frac{\alpha}{\sigma_z^2} \left| \sum_{j=1}^k y_j u_{1j}^* v_j^* \right| \right) \exp \left( -\frac{\alpha^2 \mathcal{E}_1}{\sigma_z^2} \right)}{I_0 \left( \frac{\alpha}{\sigma_z^2} \left| \sum_{j=1}^k y_j u_{0j}^* v_j^* \right| \right) \exp \left( -\frac{\alpha^2 \mathcal{E}_0}{\sigma_z^2} \right)}
}
$$

In order to proceed, we consider separately the cases where $ \alpha $ is a constant or random.

#### Unknown Frequency, Known Amplitude

In this section, we assume that $ \alpha $ is a known constant. 

The receiver structures derived in this section are similar to those found in *low-probability-of-intercept (LPI) communications*. 

In this case, the likelihood ratio above is not conditional on $ \alpha $. 

Furthermore, the last terms in the numerator and denominator are constants and can be incorporated into the threshold. 

Thus, the conditional likelihood ratio is now

$$ \boxed{
L(\vec{y}|f_d) = L_1(\vec{y}|f_d) \exp \left( -\frac{\alpha^2 (\mathcal{E}_1 - \mathcal{E}_0)}{\sigma_z^2} \right)
}
$$

where

$$
L_1(\vec{y}|f_d) = \frac{I_0 \left( \frac{\alpha}{\sigma_z^2} \left| \sum_{j=1}^k y_j u_{1j}^* v_j^* \right| \right)}{I_0 \left( \frac{\alpha}{\sigma_z^2} \left| \sum_{j=1}^k y_j u_{0j}^* v_j^* \right| \right)}
$$

In order to specify the optimum receiver, we would like to derive the unconditional likelihood ratio. 

To do this calculation, we must have knowledge of the density function of the frequency unknown, $ p(f_d) $, and we must be able to calculate expressions such as

$$
\int_{f_d} I_0 \left( \frac{\alpha}{\sigma_z^2} \left| \sum_{j=1}^k y_j u_{ij}^* v_j^* \right| \right) p(f_d) \, df_d, \quad i = 0, 1
$$

##### Approximated Likelihood Ratio

In general, we cannot solve this integral analytically; nevertheless, we can proceed using some approximation technique. 

Assume that $ f_d $ is known to be in the interval $ f_l \leq f_d \leq f_u $; that is, usually the physics of the problem allows us to estimate the maximum and minimum Doppler and frequency offsets. 

If the received frequency shift $ f_m $ occurs with probability $ P(f_m) $, then the integral above can be approximated by a sum

$$
\int_{f_l}^{f_u} I_0 \left( \frac{\alpha}{\sigma_z^2} \sum_{j=1}^k y_j u_{ij}^* v_j^* \right) p(f_d) \, df_d \approx \sum_{m=1}^M I_0 \left( \frac{\alpha}{\sigma_z^2} \sum_{j=1}^k y_j u_{ij}^* v_{jm}^* \right) P(f_m)
$$

where

$$
f_m = f_l + \frac{m - 1}{M - 1} (f_u - f_l)
$$

and

$$
v_{jm}^* = e^{-j 2\pi f_m t_j}
$$

##### Optimal Receiver
This allows us to depict an approximately optimal receiver structure as shown in Figure 6.13.

As seen in the figure, the receiver first correlates the received vector $ \vec{y} $ with the $ i $-th baseband signal shifted in frequency by $ f_m $. 

The magnitude is determined, scaled by $ \alpha / \sigma_z^2 $, and passed through a nonlinearity represented by $ I_0( ) $. 

The output is then weighted by the term $ P(f_m) $, and all $ M $ terms are summed, producing the decision variable $ U_i $.

#### Unknown Frequency, Unknown Amplitude

If the amplitude is variable and unknown, techniques similar to those developed in the previous section (of unknown amplitude) can be used. 

Specifically, we will assume that the transmitted signal energies $ \mathcal{E}_1 $ and $ \mathcal{E}_0 $ are identical and that the amplitude density function is Rician with pdf

$$
p(\alpha) = \frac{\alpha}{\sigma_r^2} \exp \left( -\frac{\alpha^2 + \kappa^2}{2 \sigma_r^2} \right) I_0 \left( \frac{\alpha \kappa}{\sigma_r^2} \right)
$$

Then we can average $p_i(\vec{y}|\alpha, f_d)$ using the density function $p(\alpha)$ and using the relationship (as we did previously)

$$
\int_0^{\infty} \alpha \exp(-c \alpha^2) I_0(g \alpha) I_0(e \alpha) \, d\alpha = \frac{1}{2c} \exp \left[ \frac{g^2 + e^2}{4c} \right] I_0 \left( \frac{g e}{2c} \right)
$$

we obtain the density function

$$
\begin{align*}
p_i(\vec{y}|f_d) 
&= \frac{1}{(2\pi \sigma_z^2)^k \sigma_r^2} \exp \left( -\frac{1}{2\sigma_z^2} \sum_{j=1}^k |y_j|^2 \right) \exp \left( -\frac{\kappa^2}{2\sigma_r^2} \right) \\
&\times \frac{1}{2c} \exp \left( \frac{\tilde{g}_i^2 + e^2}{4c} \right) I_0 \left( \frac{e}{2c} \tilde{g}_i \right)
\end{align*}
$$

where

$$
c = \frac{\mathcal{E}}{\sigma_z^2} + \frac{1}{2\sigma_r^2}
$$

the random variable $ \tilde{g}_i $ is

$$
\tilde{g}_i = \frac{1}{\sigma_z^2} \left| \sum_{j=1}^k y_j u_{ij}^* v_j^* \right|
$$

and

$$
e = \frac{\kappa}{\sigma_r^2}
$$

This allows the unconditional likelihood function to be written as

$$ \boxed{
L(\vec{y}) = \frac{\int \exp \left( \frac{1}{4c} \tilde{g}_1^2 \right) I_0 \left( \frac{e}{2c} \tilde{g}_1 \right) p(f_d) \, df_d}{\int \exp \left( \frac{1}{4c} \tilde{g}_0^2 \right) I_0 \left( \frac{e}{2c} \tilde{g}_0 \right) p(f_d) \, df_d}
}
$$

Using a similar approximation method, i.e.:

If the received frequency shift $ f_m $ occurs with probability $ P(f_m) $, then the integral above can be approximated by a sum

$$
\int_{f_l}^{f_u} I_0 \left( \frac{\alpha}{\sigma_z^2} \sum_{j=1}^k y_j u_{ij}^* v_j^* \right) p(f_d) \, df_d \approx \sum_{m=1}^M I_0 \left( \frac{\alpha}{\sigma_z^2} \sum_{j=1}^k y_j u_{ij}^* v_{jm}^* \right) P(f_m)
$$

where

$$
f_m = f_l + \frac{m - 1}{M - 1} (f_u - f_l)
$$

and

$$
v_{jm}^* = e^{-j 2\pi f_m t_j}
$$

we obtain an approximate optimal receiver structure, as shown in Figure 6.15, where

$$
\tilde{g}_{i,m} = \frac{1}{\sigma_z^2} \left| \sum_{j=1}^k y_j u_{ij}^* v_{jm}^* \right|
$$
