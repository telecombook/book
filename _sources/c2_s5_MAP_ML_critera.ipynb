## Hypothesis Testing

Consider two possible values $ s_0 $ and $ s_1 $. We wish to decide between these two alternatives when we observe a measurement $ y $ corrupted by noise $ n $. That is, we observe

$$ \mathbf{y} = s_i + \mathbf{n}, \quad i = 0, 1 $$

and decide which $ s_i $ is the most likely value. 

Most of the development of this section presumes a continuous measurement $ y $. This is merely a convenience, because $ y $ can just as easily represent discrete measurements. The theory is entirely consistent. 

In this binary case, there are two alternative hypotheses. 
Hypothesis $ H_0 $ corresponds to the value $ s_0 $ being present and hypothesis $ H_1 $ corresponds to $ s_1 $. 


## MAP Criterion

Thus, we wish to determine $ P(H_i|y) $, $ i = 0, 1 $, the probability of each of the hypotheses, given the observation $ y $. If

$$ P(H_0|y) > P(H_1|y) $$

we "decide" $ s_0 $ was present, and conversely we decide $ s_1 $ if

$$ P(H_1|y) > P(H_0|y) $$

Using Bayes' theorem, we can write

$$ P(H_0|y) = \frac{p(y|H_0)P(H_0)}{p(y)} $$

where $ P(H_i) $ is the probability of hypothesis $ H_i $ and $ p(y) $ is the pdf of $ y $; by the theorem of total probability, $ p(y) $ can be written as

$$ p(y) = p(y|H_0)P(H_0) + p(y|H_1)P(H_1) $$

Similarly,

$$ P(H_1|y) = \frac{p(y|H_1)P(H_1)}{p(y)} $$

It can be seen that

$$ p(y|H_0) = p_{\mathbf{n}}(y - s_0) $$

where $ p_{\mathbf{n}}(\cdot) $ is the noise probability density function and

$$ p(y|H_1) = p_{\mathbf{n}}(y - s_1) $$

Thus, in order to calculate the a posteriori probabilities $ P(H_i|y) $, $ i = 0, 1 $, it is necessary to know the a priori probabilities $ P(H_0) $ and $ P(H_1) $.

Let $ P(H_0) = \pi_0 $ and $ P(H_1) = \pi_1 $. From the a priori probabilities, we decide hypothesis $ H_0 $ is correct if

$$ p(y|H_0)\pi_0 > p(y|H_1)\pi_1 $$

If the inequality is reversed, we decide $ H_1 $ is correct. We will let $ D_0 $ be the event associated with the "decision" of choosing $ H_0 $ and $ D_1 $ be the corresponding event associated with $ H_1 $. Throughout the text, we will combine both inequalities in a simple expression for deciding between $ D_0 $ or $ D_1 $, i.e.,

$$ \frac{p(y|H_1)}{p(y|H_0)} \begin{array}{c} D_1 \\ \geq \\ D_0 \end{array} \frac{\pi_0}{\pi_1} $$

We define the likelihood ratio as

$$ L(y) = \frac{p(y|H_1)}{p(y|H_0)} $$

The ratio $ \frac{\pi_0}{\pi_1} $ is a fixed value $ \tau_{MAP} $ and is associated with MAP detection. 

Therefore, a MAP criterion uses the observation $ y $ to produce the likelihood ratio $ L(y) $, which is then compared to a threshold $ \tau_{MAP} $.

In the case where $ L(y) = \tau_{MAP} $, the MAP criterion offers no judgment as to which hypothesis should be chosen. Some of the other criteria presented in this chapter do address this issue. 

Since engineers abhor a "hang-up" condition where no decision is made, we shall adopt the convention of deciding $ H_1 $ when $ L(y) \geq \tau_{MAP} $.

## Maximum-Likelihood Criterion

If the two hypotheses are equally likely a priori, then $ \pi_0 = \pi_1 $ and $ \tau_{MAP} = 1 $. This special case is known as the maximum-likelihood (ML) criterion with a threshold $ \tau_{ML} = 1 $. 

It is often used in detection problems when there is no knowledge of the a priori probabilities, although in this case the analyst would best be served by first examining the minimax criterion explained in Section 4.4. It is also a powerful criterion in estimation theory, as Chapter 10 attests.

## Exmaple: Antipodal Signal Detection and Decision Process

### Setup

In this example [B2, Ex 4.1], we consider antipodal signals where $ s_0 = -b $ and $ s_1 = b $, with the a priori probabilities $ \pi_0 = 0.8 $ and $ \pi_1 = 0.2 $, respectively. We assume $ b > 0 $. 

The noise $ n $ is modeled as a zero-mean Gaussian random variable with variance $ \sigma^2 $. The probability density function (pdf) of the noise is:

$$ p_n(n) = \frac{1}{\sqrt{2\pi\sigma}} \exp\left( -\frac{n^2}{2\sigma^2} \right) $$

### Likelihood Functions

Given the observation $ y $, the likelihood functions under the two hypotheses are:

$$ p(y|H_1) = \frac{1}{\sqrt{2\pi\sigma}} \exp\left( -\frac{(y - b)^2}{2\sigma^2} \right) $$

$$ p(y|H_0) = \frac{1}{\sqrt{2\pi\sigma}} \exp\left( -\frac{(y + b)^2}{2\sigma^2} \right) $$

### Detection using MAP Criterion

To apply the Maximum A Posteriori (MAP) criterion, we need to compare the likelihood ratio $ L(y) $ to the threshold $ \tau_{MAP} $:

$$ \tau_{MAP} = \frac{\pi_0}{\pi_1} = \frac{0.8}{0.2} = 4 $$

The likelihood ratio is:

$$ L(y) = \frac{p(y|H_1)}{p(y|H_0)} = \exp\left[ \frac{(y + b)^2 - (y - b)^2}{2\sigma^2} \right] = \exp\left( \frac{2yb}{\sigma^2} \right) $$

### Decision Rule

We decide $ H_1 $ if:

$$ \exp\left( \frac{2yb}{\sigma^2} \right) \geq 4 $$

Taking the natural logarithm of both sides, we get:

$$ \frac{2yb}{\sigma^2} \geq \ln 4 $$

$$ y \geq \frac{\sigma^2 \ln 4}{2b} $$

### Numerical Example

For example, if $ b = 1 $ and $ \sigma^2 = 0.25 $, the decision rule becomes:

$$ y \geq \frac{0.25 \ln 4}{2 \cdot 1} = 0.173287 $$

#### Interpretation

If the a priori probabilities were equal ($ \pi_0 = \pi_1 = 0.5 $), then $ \tau_{MAP} $ would be 1, and the decision rule would be based purely on the likelihood ratio, i.e., $ y \geq 0 $.

### Summary of Decision Rules

- **With unequal priors ($ \pi_0 = 0.8 $, $ \pi_1 = 0.2 $)**:

  $$ y \geq 0.173287 $$

  We decide $ H_1 $ if the observation $ y $ is greater than or equal to 0.173287.

- **With equal priors ($ \pi_0 = \pi_1 = 0.5 $)**:

  $$ y \geq 0 $$
  
  We decide $ H_1 $ if the observation $ y $ is greater than or equal to 0.

This example demonstrates the application of the MAP criterion to decide between two hypotheses based on observed data, incorporating both the likelihood functions and the a priori probabilities.

### Numerical Results

```python
import numpy as np

# Given parameters
b = 1
sigma2 = 0.25
pi0 = 0.8
pi1 = 0.2

# Number of simulations
n_samples = 10000

# Generating s0 and s1
s0 = -b
s1 = b

# Generating noise
noise = np.random.normal(0, np.sqrt(sigma2), n_samples)

# Simulating received y under each hypothesis
y_given_s0 = s0 + noise
y_given_s1 = s1 + noise

# MAP criterion threshold
tau_MAP = np.log(pi0 / pi1)

# Decision rule based on MAP criterion
# Likelihood ratio test: exp(2yb / sigma2) >= pi0 / pi1
# Taking log: 2yb / sigma2 >= log(pi0 / pi1)
# Decision threshold: y >= (sigma2 / (2b)) * log(pi0 / pi1)
threshold = (sigma2 / (2 * b)) * tau_MAP

# Decisions
decisions_s0 = y_given_s0 >= threshold
decisions_s1 = y_given_s1 >= threshold

# Estimating transmitted signals (s_hat) based on decisions
s0_hat = np.where(decisions_s0, s1, s0)  # Decided s1 if decision is True, else s0
s1_hat = np.where(decisions_s1, s1, s0)

# Compute MSE for both cases
mse_s0 = np.mean((s0_hat - s0) ** 2)
mse_s1 = np.mean((s1_hat - s1) ** 2)

mse_s0, mse_s1
```

#### Plot theoretical and empirical PDFs

```python
# Re-importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

sigma = np.sqrt(sigma2)

# Empirical PDFs
bins = np.linspace(-3, 3, 1000)
empirical_pdf_H0, _ = np.histogram(y_given_s0, bins=bins, density=True)
empirical_pdf_H1, _ = np.histogram(y_given_s1, bins=bins, density=True)
y_values_mid = (bins[:-1] + bins[1:]) / 2

# Theoretical PDFs
theoretical_pdf_H0 = norm.pdf(y_values_mid, loc=s0, scale=sigma)
theoretical_pdf_H1 = norm.pdf(y_values_mid, loc=s1, scale=sigma)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(y_values_mid, empirical_pdf_H0, label='Empirical p(y|H0)', linestyle='dashed')
plt.plot(y_values_mid, empirical_pdf_H1, label='Empirical p(y|H1)', linestyle='dashed')
plt.plot(y_values_mid, theoretical_pdf_H0, label='Theoretical p(y|H0)', linestyle='solid')
plt.plot(y_values_mid, theoretical_pdf_H1, label='Theoretical p(y|H1)', linestyle='solid')

# Decision threshold line
plt.axvline(x=threshold, color='red', linestyle='--', label='Decision Threshold')

# Labels and legend
plt.xlabel('y')
plt.ylabel('Probability Density')
plt.title('Empirical and Theoretical Likelihood PDFs with Decision Threshold')
plt.legend()
plt.grid(True)
plt.show()
```

## Decision Rule

By definition, 
- $ D_0 $ is associated with the range of the observation $ y $ such that $ L(y) < \tau_{MAP} $. 
- $ D_1 $ is associated with the range of $ y $ for which $ L(y) \geq \tau_{MAP} $. 

Let $ R_0 $ and $ R_1 $ be each of these ranges of $ y $, respectively.

A decision rule $ \delta(y) $ can be defined as

$$
\delta(y) = 
\begin{cases} 
1, & y \in R_1 \\ 
0, & y \in R_0 
\end{cases}
$$

In terms of the MAP criterion, $ y $ is in $ R_1 $ if

$$
y \in R_1 \implies L(y) \geq \frac{\pi_0}{\pi_1}
$$

and

$$
y \in R_0 \implies L(y) < \frac{\pi_0}{\pi_1}
$$

It follows that the MAP decision rule is

$$
\delta_{MAP}(y) = 
\begin{cases}
1, & \forall y \ni L(y) \geq \frac{\pi_0}{\pi_1} \\
0, & \forall y \ni L(y) < \frac{\pi_0}{\pi_1}
\end{cases}
$$

## Types of Errors

Consider the case where, under hypothesis $ H_0 $, $ s_0 = 0 $. This situation is closely related to the classical radar case. A radar illuminates a specific volume of space.

The return signal can be modeled in terms of two hypotheses: 
- $ H_0 $ corresponds to "no signal" 
- $ H_1 $ corresponds to "signal present." 

In this case, $ H_0 $ is often referred to as the null hypothesis and $ H_1 $ as the alternate hypothesis.

In general, we define $ P_{ij} $ as the probability of deciding $ D_i $, when, in fact, hypothesis $ H_j $ is correct. 

In terms of the underlying probability density functions, it can be seen that

$$
P_{ij} = \int_{R_i} p(y|H_j) \, dy
$$

where the integral is understood to include the cases where the density functions are discrete or hybrid.

For binary hypotheses, some historical radar terminology is often used for three of these probabilities. 
- $ P_{01} $ is often called the probability of a "miss," $ P_m $. That is, a signal is present ($ H_1 $ is correct) and we have missed it by deciding $ D_0 $. 
- $ P_{11} $ is called the probability of detection $ P_d $. That is, we decide a signal is present and it actually is. 
- $ P_{10} $, the probability of deciding $ D_1 $ (a signal is present) when $ H_0 $ is correct (no signal present), is called the probability of a "false alarm," $ P_f $. 
- $ P_{00} $ was too mundane for radar technologists to give it a special name.

Only two of these probabilities correspond to errors, namely $ P_{01} $ and $ P_{10} $. 

In many applications (communications, for example), we are more interested in the average error probability

$$
P_E = \pi_0 P_{10} + \pi_1 P_{01}
$$

## Example: Types of Errors when using MAP decision rule with log-likelihood threshold

This example [B2, Ex 4.3] is an extension of the previous example, i.e., antipodal signal detection example. 

We define some additional terminology and observe the various decision regions for this example. 

If we define a **log-likelihood threshold** as $ \tau' $, then the four terms $ P_{ij}, i = 0, 1, j = 0, 1 $, can be written as

$$
P_{00} = \int_{-\infty}^{\tau'} \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(y + b)^2}{2\sigma^2}\right) dy
$$

$$
P_{01} = \int_{-\infty}^{\tau'} \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(y - b)^2}{2\sigma^2}\right) dy
$$

$$
P_{10} = \int_{\tau'}^{\infty} \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(y + b)^2}{2\sigma^2}\right) dy
$$

$$
P_{11} = \int_{\tau'}^{\infty} \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(y - b)^2}{2\sigma^2}\right) dy
$$

**Error function and complementary error function**

$$
\text{erf}(z) = \frac{2}{\sqrt{\pi}} \int_{0}^{z} e^{-t^2} dt
$$

and

$$
\text{erfc}(z) = \frac{2}{\sqrt{\pi}} \int_{z}^{\infty} e^{-t^2} dt = 1 - \text{erf}(z)
$$

and making the appropriate changes of variables, we can write

$$
P_{00} = \frac{1}{2} \left[ 1 + \text{erf} \left( \frac{\tau' + b}{\sqrt{2\sigma}} \right) \right]
$$

$$
P_{01} = \frac{1}{2} \left[ 1 + \text{erf} \left( \frac{\tau' - b}{\sqrt{2\sigma}} \right) \right]
$$

$$
P_{10} = \frac{1}{2} \text{erfc} \left( \frac{\tau' + b}{\sqrt{2\sigma}} \right)
$$

$$
P_{11} = \frac{1}{2} \text{erfc} \left( \frac{\tau' - b}{\sqrt{2\sigma}} \right)
$$


We can see that the MAP decision rule is

$$
\delta_{MAP}(y) = 
\begin{cases}
1, & y \geq \frac{\sigma^2}{2b} \ln 4 \\
0, & y < \frac{\sigma^2}{2b} \ln 4
\end{cases} 
$$

where the **decision regions** are

$$ R_1 = \{ y : y \geq \frac{\sigma^2}{2b} \ln 4 \} $$ 

$$ R_0 = \{ y : y < \frac{\sigma^2}{2b} \ln 4 \} $$

It follows that for the MAP criterion,

$$
\tau' = \frac{\sigma^2}{2b} \ln \tau_{MAP} = \frac{\sigma^2}{2b} \ln \frac{\pi_0}{\pi_1} = \frac{\sigma^2}{2b} \ln 4
$$

In case of the specific settings with $ b = 1 $ and $ \sigma^2 = 0.25 $, we have

$$
P_{00} \approx 0.990526
$$

$$
P_{01} \approx 0.049122
$$

$$
P_{10} \approx 0.009474
$$

$$
P_{11} \approx 0.950878
$$

```python
# Calculating empirical probabilities
P00_empirical = np.mean(s0_hat == s0)
P01_empirical = np.mean(s0_hat == s1)
P10_empirical = np.mean(s1_hat == s0)
P11_empirical = np.mean(s1_hat == s1)

# Print results
print(f"P00 (Decide H0 | H0 true): {P00_empirical:.4f}")
print(f"P01 (Decide H0 | H1 true): {P01_empirical:.4f}")
print(f"P10 (Decide H1 | H0 true): {P10_empirical:.4f}")
print(f"P11 (Decide H1 | H1 true): {P11_empirical:.4f}")
```

### Error Types for a general decision threshold

The threshold $\tau$ can represent any decision threshold, not necessarily the one specific to the MAP criterion. Here are the revised expressions with $\tau$ as the general threshold:

$ P_{00} $: Probability of deciding $ H_0 $ when $ H_0 $ is true

$$
P_{00} = \Pr(\text{Deciding } H_0 \mid H_0 \text{ is true}) = \int_{-\infty}^{\tau} p(y|H_0) \, dy
$$

$ P_{01} $: Probability of deciding $ H_0 $ when $ H_1 $ is true

$$
P_{01} = \Pr(\text{Deciding } H_0 \mid H_1 \text{ is true}) = \int_{-\infty}^{\tau} p(y|H_1) \, dy
$$

$ P_{10} $: Probability of deciding $ H_1 $ when $ H_0 $ is true

$$
P_{10} = \Pr(\text{Deciding } H_1 \mid H_0 \text{ is true}) = \int_{\tau}^{\infty} p(y|H_0) \, dy
$$

$ P_{11} $: Probability of deciding $ H_1 $ when $ H_1 $ is true

$$
P_{11} = \Pr(\text{Deciding } H_1 \mid H_1 \text{ is true}) = \int_{\tau}^{\infty} p(y|H_1) \, dy
$$

In these expressions, $\tau$ represents the decision threshold that determines the decision rule for deciding between $ H_0 $ and $ H_1 $. This threshold can vary depending on the criteria used (such as MAP, ML, Neyman-Pearson, etc.).

### Confusion matrix

The confusion matrix provides a summary of prediction results, and these elements can be used to compute various performance metrics, such as accuracy, precision, recall, and F1 score.

In the context of decision-making, particularly in binary classification, the terms $ P_{00} $, $ P_{01} $, $ P_{10} $, and $ P_{11} $ correspond to components in a confusion matrix. The confusion matrix consists of True Positives (TP), True Negatives (TN), False Positives (FP), and False Negatives (FN). Here's how they correspond:

- **$ P_{00} $**: Probability of deciding $ H_0 $ when $ H_0 $ is true.
   - **Equivalent to True Negatives (TN)**: Correctly identifying the negative class.

- **$ P_{01} $**: Probability of deciding $ H_0 $ when $ H_1 $ is true.
   - **Equivalent to False Negatives (FN)**: Incorrectly identifying the positive class as negative.

- **$ P_{10} $**: Probability of deciding $ H_1 $ when $ H_0 $ is true.
   - **Equivalent to False Positives (FP)**: Incorrectly identifying the negative class as positive.

- **$ P_{11} $**: Probability of deciding $ H_1 $ when $ H_1 $ is true.
   - **Equivalent to True Positives (TP)**: Correctly identifying the positive class.
